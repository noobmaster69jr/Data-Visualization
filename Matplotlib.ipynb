```python
x = np.arange(20)
y1 = np.linspace(1, 20, 20) + np.random.rand(20)*5
y2 = np.linspace(1, 20, 20) + np.random.rand(20)*10
plt.scatter(x, y1, s=y1*5)
plt.scatter(x, y2, s=y2*5)
plt.legend(["Shop-1", "Shop-2"])
plt.show()
```

```python
# sigmoid funtion / logistic
import math
sigmoid = lambda x : 1/(1 + math.exp(-x))
x = np.arange(-100, 100)
y = [sigmoid(i) for i in x]
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(1, 10, 0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)
fig, ax = plt.subplots(3, 1)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)
plt.show()
```

```python
x = np.arange(1, 10, 0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)
plt.figure()
plt.subplot(2, 3, 1) # indexing starts from 1 here
plt.plot(x, y1)
plt.subplot(2, 3, 3)
plt.plot(x, y2)
plt.subplot(2, 3, 5)
plt.plot(x, y3)
plt.show()
```

```python
fig = plt.figure()
ax231 = fig.add_subplot(231)
ax231.plot(x, y1)
ax233 = fig.add_subplot(233)
ax233.plot(x, y2)
ax235 = fig.add_subplot(235)
ax235.plot(x, y3)
plt.show()
```
